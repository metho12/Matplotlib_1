```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# line 
x = np.array([0,6])
y = np.array([0,200])
plt.plot(x,y)
```

```python
# points 
x = np.array([0,6])
y = np.array([0,200])
plt.plot(x,y,'o')
```

```python
# 2 polt 
x = np.array([1,2,6,8,10])
y = np.array([70,80,150,200,700])
plt.plot(x,y,'o')
```

```python
# default values
# y = default vaules (1,2,3,4,5,6,7)

x = np.array([1,2,6,8,10])
plt.plot(x)
```

```python
# specify marks

x = np.array([1,2,6,8,10])
plt.plot(x,marker='o')
```

```python
# specify marks

x = np.array([1,2,6,8,10])
plt.plot(x,marker='*')
```

```python
# specify marks

x = np.array([1,2,6,8,10])
plt.plot(x,marker='+')
```

```python
# fmt

x = np.array([1,2,6,8,10])
plt.plot(x,'o:r') # marker line color
```

```python
# fmt

x = np.array([1,2,6,8,10])
plt.plot(x,'o:r',ms=20) # marker line color
```

```python
# border 
x = np.array([1,2,6,8,10])
plt.plot(x,'o:r',ms=20,mec='g') # marker line color
```

```python
# marker 
x = np.array([1,2,6,8,10])
plt.plot(x,'o:r',ms=20,mfc='g') # marker line color
```

```python
# marker 
x = np.array([1,2,6,8,10])
plt.plot(x,'o:r',ms=20,mfc='#2e2e2e') 
```

```python
# line style  
x = np.array([1,2,6,8,10])
plt.plot(x,ms=20,mfc='#2e2e2e',linestyle='dashed') 
```

```python
# line style  
x = np.array([1,2,6,8,10])
plt.plot(x,ms=20,mfc='#2e2e2e',linestyle='dashed',color='y',linewidth=10) 
```

```python
# line style  
x = np.array([1,2,6,8,10])
plt.plot(x,ms=20,mfc='#2e2e2e',color='y',linewidth=10) 
```

```python
# multiple line 
x = np.array([0,6])
y = np.array([0,200])
plt.plot(x)
plt.plot(y)
```

```python
# label 
x = np.array([0,6])
y = np.array([0,200])
plt.plot(x)
plt.plot(y)

plt.xlabel('sales')
plt.ylabel('months')
plt.title('main title')
```

```python
# label 
x = np.array([0,6])
y = np.array([0,200])
plt.plot(x)
plt.plot(y)

plt.xlabel('sales',loc='left')
plt.ylabel('months')
plt.title('main title')
```

###  fonts 

```python
font1 = {'family':'impact','color':'blue','size':15}

x = np.array([0,6])
y = np.array([0,200])
plt.plot(x)
plt.plot(y)

plt.xlabel('sales',loc='left',fontdict=font1)
plt.ylabel('months')
plt.title('main title')
```
